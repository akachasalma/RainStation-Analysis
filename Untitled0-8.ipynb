````python
import sklearn
#pour standardisation (centrage et reduction)
from sklearn.preprocessing import StandardScaler

# Sélection des colonnes numériques pour l'ACP
data_numeric = dt[['x_utm', 'y_utm', 'zone_utm', 'Longitude', 'Latitude']]

sc = StandardScaler()
#transformation – centrage-réduction
Z = sc.fit_transform(data_numeric)
print(Z)
````

````python
Coor = data_numeric.corr()
print(Coor)
````

````python
Coor = data_numeric.drop(columns=['zone_utm']).corr()
print(Coor)
````

````python
import matplotlib.pyplot as plt
import seaborn as sns

plt.subplots(figsize=(10,4))
sns.heatmap(data_numeric.corr())
plt.show()
````

````python
eigenvalues, eigenvectors = np.linalg.eig(Coor)

diagonal_matrix = np.diag(eigenvalues)

#affichage des valeurs propres en ordre décroissant
print(diagonal_matrix)
````

````python
# Calcul du pourcentage de variance expliquée par chacun des axes factoriels
tot = sum(eigenvalues)
var_exp = [(i / tot)*100 for i in sorted(eigenvalues, reverse=True)]
cum_var_exp = np.cumsum(var_exp)
print(pd.DataFrame({'valprop':eigenvalues,'inertie':var_exp,'inertiecum':cum_var_exp}))
````

````python
import matplotlib.pyplot as plt
import numpy as np


# Diagramme des valeurs propres
plt.figure(figsize=(10, 5))
plt.plot(eigenvalues, '-o')
plt.title('Diagramme des valeurs propres')
plt.xlabel('Indice')
plt.ylabel('Valeur propre')
plt.grid(True)
plt.show()

# Calcul des pourcentages cumulés
cumulative_percentages = np.cumsum(eigenvalues) / np.sum(eigenvalues) * 100

# Diagramme des pourcentages cumulés
plt.figure(figsize=(10, 5))
plt.plot(cumulative_percentages, '-o')
plt.title('Diagramme des pourcentages cumulés des valeurs propres')
plt.xlabel('Indice')
plt.ylabel('Pourcentage cumulé (%)')
plt.grid(True)
plt.show()
````

````python
from sklearn.decomposition import PCA
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# DataFrame X
# scaler = StandardScaler()
# Z = scaler.fit_transform(X)

# Initialisation et ajustement de PCA sur les données standardisées Z
pca = PCA(n_components=2)
pca.fit(Z)

# Récupération des vecteurs propres pour les deux premières composantes
selected_eigenvectors = pca.components_.T

# Projection des données sur les deux premières composantes
Tp = np.dot(Z, selected_eigenvectors)

# Visualisation des individus sur les deux premiers axes
plt.figure(figsize=(6, 4))
plt.scatter(Tp[:, 0], Tp[:, 1], color='blue')
plt.xlabel('Axe factoriel 1')
plt.ylabel('Axe factoriel 2')
plt.title('Projection des individus sur les deux premiers axes factoriels')
plt.grid(True)
plt.show()

# Calcul des contributions des variables aux composantes
# Carré des coordonnées des vecteurs propres (loadings)
loadings_squared = selected_eigenvectors ** 2
# Pour chaque composante, division par la somme pour obtenir la contribution relative
contributions = loadings_squared / np.sum(loadings_squared, axis=0)

# Affichage des contributions en DataFrame
contributions_df = pd.DataFrame(contributions, index=data_numeric.columns, columns=['PC1_contrib', 'PC2_contrib'])
print("Contributions des variables aux deux premières composantes :")
print(contributions_df)

# Visualisation des contributions
for i in range(len(data_numeric.columns)):
    plt.arrow(0, 0, selected_eigenvectors[i, 0], selected_eigenvectors[i, 1], color='red', width=0.01, head_width=0.05)
    plt.text(selected_eigenvectors[i, 0], selected_eigenvectors[i, 1], data_numeric.columns[i], color='red', ha='right')
plt.xlabel('Axe factoriel 1')
plt.ylabel('Axe factoriel 2')
plt.title('Contribution des variables aux deux premiers axes factoriels')
plt.grid(True)
plt.show()
````

````python
# Contributions des individus
# Sur un axe j, c'est le carré de la coordonnée divisé par la valeur propre associée à cet axe.
contributions_ind = (Tp ** 2) / np.sum(Tp ** 2, axis=0)

# Conversion en DataFrame pour une meilleure lisibilité
contributions_ind_df = pd.DataFrame(contributions_ind, columns=['PC1_contrib', 'PC2_contrib'])
print(contributions_ind_df)

````

````python
# Qualité de la représentation des individus (cos2)
cos2_ind = Tp ** 2
cos2_ind /= np.sum(Tp ** 2, axis=1)[:, None]

# Conversion en DataFrame
cos2_ind_df = pd.DataFrame(cos2_ind, columns=['PC1_cos2', 'PC2_cos2'])
print(cos2_ind_df)


````

````python
# Qualité de représentation des variables (cos2)
cos2_var = loadings_squared
# Somme des carrés des loadings pour chaque variable (somme sur les composantes)
sum_loadings_squared = np.sum(loadings_squared, axis=1)

# cos2 pour chaque variable sur chaque composante
cos2_var = loadings_squared / sum_loadings_squared[:, None]

# cos2_var est dans la bonne forme (nombre de variables, nombre de composantes)
cos2_var_df = pd.DataFrame(cos2_var, index=data_numeric.columns, columns=['PC1_cos2', 'PC2_cos2'])
print(cos2_var_df)

````

````python
from scipy.cluster.hierarchy import linkage
import matplotlib.pyplot as plt

# data_numeric est la DataFrame de données numériques
# Calcul de la matrice de linkage
Z = linkage(data_numeric, 'ward')

# Tri des hauteurs de fusion (inerties) par ordre décroissant
inertias = sorted(Z[:, 2], reverse=True)

# Tracé des 20 plus grands sauts d'inertie
plt.figure(figsize=(5, 3))
plt.plot(range(1, 21), inertias[:20], drawstyle="steps-post")
plt.xlabel('Nombre de classes')
plt.ylabel('Inertie')
plt.title('Les 20 plus grands sauts d’inertie dans le dendrogramme')
plt.xticks(range(1, 21))  # l'axe des x a les bonnes étiquettes
plt.grid(True)  # Ajout d'une grille pour une meilleure lisibilité
plt.show()
````

````python
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Normalisation des données
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data_numeric)

# Génération du lien hiérarchique
Z = linkage(data_scaled, method='ward', metric='euclidean')

# Affichage de dendrogramme
plt.figure(figsize=(10, 6))
plt.title('Dendrogramme Hiérarchique')
plt.xlabel('Index des Échantillons')
plt.ylabel('Distance')
dendrogram(Z)
plt.show()

# Affichage d'inertie du dendrogramme (méthode du coude)
last_p_merge_distances = Z[-20:, 2]
last_p_merge_distances_reversed = last_p_merge_distances[::-1]

plt.figure(figsize=(8, 6))
plt.title('Inertie du Dendrogramme')
plt.xlabel('Nombre de Clusters')
plt.ylabel('Inertie (Distance de Fusion)')
plt.plot(range(1, 21), last_p_merge_distances_reversed, marker='s')
plt.xticks(range(1, 21))
plt.show()

# Détermination du nombre de clusters (basé sur le graphe précédent)
k = int(input("Entrez le nombre de clusters à retenir : "))

# Découpage du dendrogramme pour former les clusters
clusters = fcluster(Z, k, criterion='maxclust')
data_numeric['Cluster'] = clusters

# Analyse des clusters
print(data_numeric.groupby('Cluster').mean())

# Réalisation d'une ACP pour visualiser les clusters (si les données sont prêtes pour une ACP)
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(data_scaled)
principalDf = pd.DataFrame(data=principalComponents, columns=['PC1', 'PC2'])
principalDf['Cluster'] = clusters

# Visualisation des clusters sur les deux premiers composants principaux
plt.figure(figsize=(8, 6))
for i in np.unique(clusters):
    plt.scatter(principalDf[principalDf['Cluster'] == i]['PC1'],
                principalDf[principalDf['Cluster'] == i]['PC2'],
                label=f'Cluster {i}')
plt.title('Visualisation des clusters CAH sur les composantes principales')
plt.xlabel('Premier Composant Principal')
plt.ylabel('Deuxième Composant Principal')
plt.legend()
plt.show()
````

````python
import pandas as pd
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
import numpy as np


# Génération de la matrice de linkage pour CAH
Z = linkage(data_numeric, method='ward', metric='euclidean')

# Choix du nombre de clusters basé sur votre analyse dendrogramme
n_clusters = 5

cah_labels = fcluster(Z, n_clusters, criterion='maxclust')

# Tri des index en fonction des étiquettes de clusters obtenues
idc = np.argsort(cah_labels)

# Affichage des observations et leurs groupes
print(pd.DataFrame(data_numeric.index[idc], cah_labels[idc]))

# 'clusters' contient une étiquette de cluster pour chaque observation dans 'data_numeric'
V_Labels = dt['region_fr']

# Correspondance avec les vrais labels
crosstab_cah = pd.crosstab(V_Labels, cah_labels)
print(crosstab_cah)
````

````python
from sklearn.metrics.cluster import v_measure_score
# Récupération des étiquettes de cluster pour chaque point de données
v_measure_cah = v_measure_score(V_Labels, cah_labels)

print(f"Le taux de bon classement du modèle CAH sur la variable 'region_fr' est de {v_measure_cah:.2f}")
````
